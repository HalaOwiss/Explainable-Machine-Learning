```python
fuel_samples2 = fuel.sample(n=500)
X2 = fuel_samples2[['co2TailpipeGpm','Year','displ','pv4','cylinders','lv4','hlv','ghgScore','co2']]
Y2 = fuel_samples2[['comb08']]
```

```python
X2.isna().sum()
```

```python
X2 = X2.fillna(X2.mean())
```

```python
rand = 9
X_train, X_test, y_train, y_test = train_test_split(X2, Y2,\
test_size=0.15, random_state=rand)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train,\
test_size=0.2, random_state=rand)
```

```python
y_val.info()
```

```python
X_train.shape[1]
```

```python
import tensorflow as tf
from tensorflow.keras import layers


# Build the neural network
model = tf.keras.Sequential([
    # Preprocessing layer for normalizing continuous features
    layers.InputLayer(input_shape=(X_train.shape[1],)),
    layers.experimental.preprocessing.Normalization(),

    # Hidden layer with 64 neurons and ReLU activation
    layers.Dense(64, activation='relu'),
    layers.Dense(64, activation='relu'),

    # Output layer for 1 output
    layers.Dense(1)
])

# Compile the model
model.compile(optimizer=tf.keras.optimizers.Adam(lr = 0.0005), loss='mean_squared_error',metrics=['mse'])

print('summry ',model.summary())

es = tf.keras.callbacks.EarlyStopping(monitor='val_loss',mode = 'min',verbose =1, patience=200,min_delta = 0.0005, restore_best_weights=True)
# Train the model
fitted_model=model.fit(X_train.astype(float), y_train.astype(float), epochs=3000, batch_size=128, validation_data=(X_val.values.astype(float), y_val.values.astype(float)),verbose=1,callbacks=[es])
```

```python
from plot_keras_history import show_history,plot_history

show_history(fitted_model)
plot_history(fitted_model,path = "standard.png")
plt.close()
```

```python
y_train_nn_pred = model.predict(X_train.astype(float))
y_test_nn_pred = model.predict(X_test.astype(float))
```

```python
from sklearn.metrics import r2_score
# Compute R-squared
r_squared = r2_score(y_test, y_test_nn_pred)

print(f'R-squared test: {r_squared}')
r_squared = r2_score(y_train, y_train_nn_pred)

print(f'R-squared train: {r_squared}')
```

```python
from sklearn.tree import DecisionTreeRegressor
fitted_dt_surrogate = DecisionTreeRegressor(max_depth=7,\
random_state=9).fit(X_train, y_train_nn_pred)
y_train_dt_pred = fitted_dt_surrogate.predict(X_train)
y_test_dt_pred = fitted_dt_surrogate.predict(X_test)
```

```python
RMSE_dt_nn_train = mean_squared_error(y_train_nn_pred,\
y_train_dt_pred,\
squared=False)
RMSE_dt_nn_test = mean_squared_error(y_test_nn_pred,\
y_test_dt_pred,\
squared=False)
R2_dt_nn_test = r2_score(y_test_nn_pred, y_test_dt_pred)
#Print all metrics
print('RMSE_train: %.4f\tRMSE_test: %.4f\tr2: %.4f' %\
(RMSE_dt_nn_train, RMSE_dt_nn_test, R2_dt_nn_test))
```

```python
from rulefit import RuleFit
fitted_rf_surrogate = RuleFit(max_rules=150, rfmode='regress',\
random_state=rand, tree_size=8).\
fit(X_train.astype(float).values,\
np.array(y_train_nn_pred).squeeze(),\
X_train.columns)
```

```python
y_train_rf_pred = fitted_rf_surrogate.predict(X_train.astype(float).values)
y_test_rf_pred = fitted_rf_surrogate.predict(X_test.astype(float).values)
```

```python
#Measure how well Rule Fit replicates Neural Network's predictions
RMSE_rf_nn_train = mean_squared_error(y_train_nn_pred,\
y_train_rf_pred,\
squared=False)
RMSE_rf_nn_test = mean_squared_error(y_test_nn_pred,\
y_test_rf_pred,\
squared=False)
R2_rf_nn_test = r2_score(y_test_nn_pred, y_test_rf_pred)
#Print all metrics
print('RMSE_train: %.4f\tRMSE_test: %.4f\tr2: %.4f' %\
(RMSE_rf_nn_train, RMSE_rf_nn_test, R2_rf_nn_test))
```
